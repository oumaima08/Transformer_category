```python
import pandas as pd

# Exemple de DataFrame (colonne cat√©gorie)
data = {
    'product_id': [1, 2, 3, 4, 5],
    'product_name': ['Laptop', 'Smartphone', 'Desk', 'Chair', 'Headphones'],
    'category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Electronics'],
    'price': [1200.50, 799.99, 300.00, 150.75, 199.99]
}

df= pd.DataFrame(data)
print("DataFrame :")
print(df)
```

```python
# Encodage binaire
df['category'] = df['category'].map({'Electronics': 0, 'Furniture': 1})
df.head()
```

```python
# Encodage one-Hot
print(df['product_name'].dtype)
df['product_name'] = df['product_name'].astype('category')
df_encoded = pd.get_dummies(df, columns=['product_name', ], dtype=int)
print(df_encoded.iloc[0,:])
df_encoded.head()
```
